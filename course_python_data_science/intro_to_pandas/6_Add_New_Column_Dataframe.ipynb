# Adding a new column with an array

El dataframe tiene 1000 filas, por lo que necesitamos crear una matriz con 1000 elementos

```python
# import numpy
import numpy as np
```

```python
# crear una matriz de 1000 elementos
language_score = np.arange(0, 1000)
```

```python
# longitud de la matriz
len(language_score)
```

```python
# agregando una nueva columna al dataframe con una matriz
df_exams["language score"] = language_score
df_exams
```

```python
# crear números enteros aleatorios entre 1 y 100
int_language_score = np.random.randint(1, 100, size=1000)
```

```python
# valor mínimo incluido y valor alto exclusivo
print(min(int_language_score))
print(max(int_language_score))
```

```python
# agregando una nueva columna al dataframe con una matriz
df_exams["language score"] = int_language_score
df_exams
```
