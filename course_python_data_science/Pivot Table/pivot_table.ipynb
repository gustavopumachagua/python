# pivot_table()

```python
# leer y mostrar el conjunto de datos
df_sales = pd.read_excel("./supermarket_sales.xlsx")
df_sales
```

```python
# hacer una tabla dinÃ¡mica y agregar una funciÃ³n agregada
df_sales.pivot_table(index="Gender", aggfunc="sum").round(1)
```

```python
# haga una tabla dinÃ¡mica, agregue una funciÃ³n agregada y seleccione algunas columnas
df_sales.pivot_table(index="Gender", values=["Quantity", "Total"], aggfunc="sum").round(1)
```

```python
# haga una tabla dinÃ¡mica que diga cuÃ¡nto gastan hombres y mujeres en cada categorÃ­a
df_sales.pivot_table(index="Gender", columns="Product line", values="Total", aggfunc="sum").round(1)
```
