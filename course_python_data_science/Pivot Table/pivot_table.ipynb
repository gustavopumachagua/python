```python
import pandas as pd
```

# pivot()

```python
# leer y mostrar el conjunto de datos
df_gdp = pd.read_csv('./gdp.csv')
df_gdp
```

```python
# remodelar el dataframe con .pivot()
df_gdp.pivot(index="year", columns="country", values="gdppc")
```

# pivot_table()

```python
# leer y mostrar el conjunto de datos
df_sales = pd.read_excel("./supermarket_sales.xlsx")
df_sales
```

```python
# hacer una tabla dinámica y agregar una función agregada
df_sales.pivot_table(index="Gender", aggfunc="sum").round(1)
```

```python
# haga una tabla dinámica, agregue una función agregada y seleccione algunas columnas
df_sales.pivot_table(index="Gender", values=["Quantity", "Total"], aggfunc="sum").round(1)
```

```python
# haga una tabla dinámica que diga cuánto gastan hombres y mujeres en cada categoría
df_sales.pivot_table(index="Gender", columns="Product line", values="Total", aggfunc="sum").round(1)
```
