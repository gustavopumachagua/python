```python
# leyendo el archivo csv
df_population_raw = pd.read_csv("./population_total.csv")
```

```python
# primer vistazo al conjunto de datos
df_population_raw
```

# Making a Pivot Table


.pivot (): devuelve el dataframe remodelado organizado por valores de índice / columna dados ("pivote sin agregación")

```python
# descartar valores nulos
df_population_raw.dropna(inplace=True)
```

```python
# hacer una tabla dinámica
df_pivot = df_population_raw.pivot(index="year", columns="country", values="population")
```

```python
# seleccionando algunos paises
df_pivot = df_pivot[["United States", "India", "China", "Indonesia", "Brazil"]]
```

```python
# mostrar pivot table
df_pivot
```

```python
# importando matplotlib
import matplotlib.pyplot as plt
```

# Lineplot

```python
df_pivot.plot(kind="line", xlabel="Years", ylabel="Population", title="Population (1995-2020)", figsize=(8, 4))

# save plot
plt.savefig("my_test.png")

# show plot
plt.show()
```
